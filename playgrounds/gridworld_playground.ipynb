### GridWorld Playground
Playground used to experiment with different algorithms on the custom made GridWorld environment.

```python
import sys
sys.path.append('/Users/bogdanfloris/Workspace/Code/Reinforcement-Learning/')
from library.dynamic_programming import dynamic_programming as dp
from library.td_learning import temporal_diff_learning as td
from library.environments.gridworld import GridWorldEnv
from library.plotting import plot_episode_stats
```

Define the environment

```python
env = GridWorldEnv()
```

#### Dynamic Programming
Experiments on the GridWorld environment using dynamic programming algorithms

```python
policy, _ = dp.policy_iteration(env=env)
```

```python
print("Optimal policy found using Policy Iteration algorithm")
env.render_policy(policy=policy)
```

#### Temporal Difference Learning
Experiments on the GridWorld environment using temporal difference learning algorithms

```python
# SARSA hyperparameters
num_episodes = 1000
epsilon = 0.1
```

```python
q, _ = td.sarsa(env, num_episodes, epsilon=epsilon)
policy = td.make_epsilon_greedy_policy(epsilon=0.0, action_count=env.action_space.n, q=q)
print("Policy after {} episodes of {} for \u03B5 = {}".format(num_episodes, 'SARSA', epsilon))
env.render_policy(policy=policy)
```

```python
# Q Learning hyperparameters
num_episodes = 1000
epsilon = 0.1
```

```python
q, stats = td.q_learning(env, num_episodes, epsilon=epsilon)
policy = td.make_epsilon_greedy_policy(epsilon=0.0, action_count=env.action_space.n, q=q)
print("Policy after {} episodes of {} for \u03B5 = {}".format(num_episodes, 'Q-Learning', epsilon))
env.render_policy(policy=policy)
```
