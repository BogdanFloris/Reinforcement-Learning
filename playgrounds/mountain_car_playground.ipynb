### Mountain Car Playground
Playground for the Mountain Car OpenAI gym environment used to demonstrate the Q-Learning algorithm with Function Approximation (Stochastic Gradient Descent)

```python
import sys
import gym
sys.path.append('../')
from library.td_learning import temporal_diff_learning as td
from library.estimators.estimators import QFunctionSGD
from library.plotting import plot_cost_to_go_mountain_car, plot_episode_stats
```

```python
# initialize the environment
env = gym.make('MountainCar-v0')
```

```python
# initialize estimator
estimator = QFunctionSGD(env)
```

```python
# run Q learning
_, stats = td.q_learning(env, num_episodes=100, estimator=estimator, epsilon=0.0)
```

```python
# plot cost function and statistics
plot_cost_to_go_mountain_car(env, estimator)
plot_episode_stats(stats, smoothing_window=25)
```

```python
# initialize estimator
estimator = QFunctionSGD(env)
# run Q learning with 1000 episodes
_, _ = td.q_learning(env, num_episodes=1000, estimator=estimator, epsilon=0.0)
```
